```python
# General to handle data
import pandas as pd
import datetime
import numpy as np

# For plotting and visualization
import matplotlib.pyplot as plt
import seaborn as sns

#palette = sns.color_palette("pastel")
#sns.palplot(palette)

#sns.set_style('whitegrid')
#plt.style.use('seaborn-pastel')
```

```python
# Load the confirmed cases df
global_confirmed_cases_df = pd.read_csv ('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
global_confirmed_cases_df.sort_values (by = ['Country/Region'])
```

```python
# Load the deaths cases df
global_deaths_cases_df = pd.read_csv ('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
global_deaths_cases_df.sort_values (by = ['Country/Region'])
```

```python
# Load the recoveries cases df
global_recoveries_cases_df = pd.read_csv ('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
global_recoveries_cases_df.sort_values (by = ['Country/Region'])
```

```python
# Dataframes modification - Organize the dates within a dedicated column (Date) & Create Month-Year column 
def dataframes_modification (dfs, titles):
    dfs = dfs.melt(
    id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'],  
    var_name = 'Date', 
    value_name = titles
    ).sort_values (by = ['Country/Region'])
    
    #df['Date'] = df['Date'].astype('datetime64[ns]').dt.strftime('%m-%B-%Y')
    dfs['Month-Year'] = dfs['Date'].astype('datetime64[ns]').dt.strftime('%B-%Y')
    dfs = dfs.rename(columns = {'Country/Region': 'Country'}, inplace = False)
    return dfs
```

```python
# Check if the modify has been applied to the dataframes
global_confirmed_cases_df = dataframes_modification (global_confirmed_cases_df, 'Morbidity rate over time')

global_confirmed_cases_df
```

```python
global_deaths_cases_df = dataframes_modification (global_deaths_cases_df, 'Mortality rate over time')

global_deaths_cases_df
```

```python
global_recoveries_cases_df = dataframes_modification (global_recoveries_cases_df, 'Recovery rate over time')

global_recoveries_cases_df
```

```python
global_confirmed_cases_df.info()
global_confirmed_cases_df.isnull().sum()
```

```python
# Clean dataframes
def clean_columns_headline (df):
    df.columns = df.columns.str.strip ()

def clean_blank_rows (df):
    df.dropna (how = 'all', inplace = True)
    
def set_columns_type (df):
    df['Date'] = pd.to_datetime(df['Date'])
    
def replace_missing_values_with_default (df):
    df['Province/State'].fillna('',inplace = True)
    df['Country'].fillna('',inplace = True)
    
def clean_file (df):
    clean_columns_headline (df)
    clean_blank_rows (df)
    set_columns_type (df)
    replace_missing_values_with_default (df)
```

```python
clean_file (global_confirmed_cases_df)

global_confirmed_cases_df.info()
global_confirmed_cases_df.isnull().sum()
```

```python
clean_file (global_deaths_cases_df)
```

```python
clean_file (global_recoveries_cases_df)
```

```python
def average_rate_over_time (dfs, titles):
    dfs = dfs.groupby('Month-Year')[titles].mean().sort_values()
        
    sns.set(rc = {'figure.figsize':(12,6)})
    axes = sns.lineplot(data = dfs)
        
    axes.set_title(titles)
    axes.set_xlabel("Date")
    axes.set_ylabel("# of Cases")
    
    plt.xticks(rotation = 40)
    plt.legend(labels = [titles, titles, titles])
    #plt.show()
    
    #return axes
```

```python
# International morbidity average rate over time plot
average_rate_over_time (global_confirmed_cases_df, 'Morbidity rate over time')
```

#### 5. Total morbidity rate by country - USA, Israel, Italy, Spain, from the beginning of the corona (January 1st 2020)

```python
def specific_countries_average_rate_over_time (dfs, titles):
    dfs = dfs.loc[dfs.Country.isin(['US', 'Israel', 'Italy', 'Spain']), :]
    dfs = dfs.groupby('Country')[[titles]].sum()
    #dfs['Country'] = dfs.index
    
    sns.set(rc = {'figure.figsize':(12,7)})
    axes = sns.catplot(x = dfs.index, y = titles, kind = "bar", data = dfs)
    
    axes.fig.suptitle(titles)
    #plt.show()
    
    #return axes
```

```python
# Total morbidity rate by country plot
specific_countries_average_rate_over_time (global_confirmed_cases_df, 'Morbidity rate over time')
```

#### 6. Total mortality rate by country - USA, Israel, Italy, Spain, from the beginning of the corona (January 1st 2020)

```python
# Total mortality rate by country plot
specific_countries_average_rate_over_time (global_deaths_cases_df, 'Mortality rate over time')
```

#### 7. Total recovery rate by country - USA, Israel, Italy, Spain, from the beginning of the corona (January 1st 2020)

```python
# Total recovery rate by country plot
specific_countries_average_rate_over_time (global_recoveries_cases_df, 'Recovery rate over time')
```

#### 8. Average rate by country - USA, Israel, Italy, Spain, from the beginning of the corona (January 1st 2020) over time - Comparison

```python
countries_to_compare = ['US', 'Israel', 'Italy', 'Spain']
    
def countries_comparison (dfs ,countries_to_compare, titles):
    month_dict = {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 
              'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
    
    # Split Month-Year (str), map it according to MM-YEAR (int) oin order to plot the accurate comparison in Month-Year
    for i in range(len (countries_to_compare)):
        df_comparison = dfs[dfs['Country'] == countries_to_compare[i]].groupby('Month-Year')[[titles]].mean()
        df_comparison['Date'] = df_comparison.index
        split =  df_comparison['Date'].str.split('-', expand = True)
        df_comparison['Month'], df_comparison['Year'] = split[0], split[1]
        df_comparison['Month'] = df_comparison['Month'].map(month_dict)
        df_comparison['Year'] = df_comparison['Year'].map({'2020':1,'2021':2})
        df_comparison = df_comparison.sort_values(['Year','Month'])[titles]
        
        sns.set(rc = {'figure.figsize':(12,7)})
        axes = sns.lineplot(data = df_comparison)
        
    axes.set_title(titles)
    axes.set_xlabel("Date")
    axes.set_ylabel("# of Cases")
    plt.xticks(rotation = 40)
    plt.legend(countries_to_compare)
    #plt.show()
    
    #return axes
    
```

```python
# Countries Comparison plot
countries_comparison (global_confirmed_cases_df ,countries_to_compare, 'Morbidity rate over time')
```

```python
countries_comparison (global_deaths_cases_df, countries_to_compare, 'Mortality rate over time')
```

```python
countries_comparison (global_recoveries_cases_df, countries_to_compare, 'Recovery rate over time')
```
